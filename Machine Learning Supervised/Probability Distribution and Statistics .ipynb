# Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate a normal distribution (1-D)

u=5 # if u=0 sigma=1 then std normal distribution
sigma=1
#such a distribution is also called std normal distribution

vals=u+sigma*np.random.randn(1000)
print(vals.shape)

plt.hist(vals,100,ec='white') #no of buckets=100
plt.show()

#This is depicting how many values fall into each interval
```

```python
vals=np.round(vals)#round of to int
z=np.unique(vals,return_counts=True)
print(z)
#numbs closer to 2. are 6
```

```python
u=5
sigma=0.2 #spread


vals=u+sigma*np.random.randn(200)
print(vals.shape)
#print(vals)

plt.hist(vals,10,ec='white') #no of buckets=100
plt.show()
```

```python
x= vals
y= np.zeros(x.shape)
plt.scatter(x,y)
```

# Multivariate Normal Distribution 

```python
#BiVariate 
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])

dist=np.random.multivariate_normal(mean,cov,500) #distribution points X matrix in register the feature matrix

print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])

#sigmaxx=1 sigmayy=1 sigmaxy=sigmayx=0 that is x and y are independent
```

```python
#tilted distribution along the line y=x more dependence more tilt
mean=np.array([0.0,0.0])
cov=np.array([[1,0.8],[0.8,1]])

mean2=np.array([5.0,6.0])
cov2=np.array([[1.3,0.8],[0.8,1.1]]) #spread more in x and y that in cov

dist=np.random.multivariate_normal(mean,cov,500) #distribution points X matrix in register the feature matrix
dist2=np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1],color='orange')
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
