```python
import cv2
```

```python
img = cv2.imread("Sketchpad.png")

newImg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #back to RGB from BGR


```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img) #R,G,B channels in general, when we read through mathplotlib it thinks B,G,R by default
plt.show()
plt.imshow(newImg)
plt.show()
```
